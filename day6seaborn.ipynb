```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
##matplot libary tools for vitualization 
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
plt.style.use("ggplot")
sns.displot(x=df["total_bill"],kde=True)
```

```python
sns.regplot(x=df["total_bill"],y=df["tip"])
```

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"])
```

```python
sns.countplot(x=df["day"],color="black")
```

```python
plt.figure(figsize=(10,8))
ax= sns.barplot(x="sex",y="total_bill",data=df,palette="Blues", estimator="max",hue="day")
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x()+ p.get_width()/2.,p.get_height()),
               ha="center",va="baseline",fontsize=10,color="magenta",xytext=(0,5),
               textcoords="offset points")
plt.title("day Distribution")
plt.xticks(rotation=45)
plt.legend(loc=[1,0])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(x="sex",y="total_bill",data=df,palette="Blues",hue="day")
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x="sex",y="total_bill",data=df,palette="Blues",hue="day")
```

```python
plt.figure(figsize=(10,8))
sns.jointplot(x="total_bill",y="tip",data=df)
```

```python
sns.pairplot(df)
```
