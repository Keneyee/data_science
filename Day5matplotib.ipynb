```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("RealGDP20122023.csv")
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.rename({'index':'Year', 'tyear':'Period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Livestock',
       'Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MininingQua', 'MiningAndQuarrying':'CrudeOil',
       'CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'Minerals',
       'QuarryingAndOtherMinerals':'Manufacturing','Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBev',
       'FoodBeverageAndTobacco':'Textile', 'TextileApparelAndFootwear':'WoodProduct',
       'WoodAndWoodProducts':'PaperProduct', 'PulpPaperAndPaperProducts':'PhamaProduct',
       'ChemicalAndPharmaceuticalProducts':'NonMetalPrd', 'NonMetallicProducts':'PlasticProduct',
       'PlasticAndRubberProducts':'EEE', 'ElectricalAndElectronics':'IronSteel',
       'BasicMetalIronAndSteel':'VehiclesASS', 'MotorVehiclesAndAssembly':'OtherManu',
       'OtherManufacturing':'Gasstem', 'ElectricityGasSteamAndAirCon':'Watersupply',
       'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'Accommodation',
       'AccommodationAndFoodServices':'Transport', 'TransportationAndStorage':'RoadTransport',
       'RoadTransport':'Rail', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport',
       'AirTransport':'TransService', 'TransportServices':'PostServices', 'PostAndCourierServices':'ICT',
       'InformationAndCommunication':'Publishing', 'Publishing':'MusicProduction',
       'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'Entertainment',
       'ArtsEntertainmentAndRecreation':'FinInsurance', 'FinanceAndInsurance':'FinInstitution',
       'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'SciTech',
       'ProfessionalScientificAndTechnicalServices':'Adminstration',
       'AdministrativeAndSupportServicesBusinessServices':'PublicAdminstration',
       'PublicAdministration':'Education', 'Education':'HealthService', 'HumanHealthAndSocialServices':'OtherService',
       'OtherServices':'GDP2010', 'GDP at 2010 Constant Basic Prices':'Nettaxs',
       'NetTaxesOnProducts':'GDP20102', 'GDP at 2010 Constant Market Prices':'NILL'},axis = 1,inplace =True)
df.head()
```

```python
df.drop("NILL",axis=1,inplace=True)
```

```python
df.head
```

```python
plt.plot(df["Year"],df["CrudeOil"])
```

```python
df["Period"].unique()
```

```python
df2=df[df["Period"]=="Annual"]
```

```python
plt.style.use("ggplot")
plt.plot(df2["Year"],df2["CrudeOil"],ls="--",lw=2.5,marker="o",mec="BLUE",mfc="BLUE")
plt.xlabel("Year")
plt.ylabel("CrudeOil")
plt.title("trend of crude oil gdp in millions")
plt.show()
```

```python
df.columns
```

```python
figure=plt.figure(figsize=(10,10))
axes=figure.add_axes([0.1,0.1,0.8,0.7]) ## left,button, width and height
axes.plot(df2["Year"],df2["CrudeOil"],ls= "--",lw=2.5,marker="o",mec="BLUE",mfc="BLUE",color="black",markevery=3,label="Crudeoil")
axes.plot(df2["Year"],df2["Agriculture"],ls= "--",lw=2.5,marker="o",mec="BLUE",mfc="BLUE",color="green",markevery=3,label="Agric")
axes.plot(df2["Year"],df2["ICT"],ls= "--",lw=2.5,marker="o",mec="BLUE",mfc="BLUE",color="blue", markevery=3,label="ICT")
axes.plot(df2["Year"],df2["Education"],ls= "--",lw=2.5,marker="o",mec="BLUE",mfc="BLUE",color="magenta",markevery=3,label="EDUC")
axes.plot(df2["Year"],df2["Trade"],ls= "--",lw=2.5,marker="o",mec="BLUE",mfc="BLUE",color="red",markevery=3,label="TRADE")
axes.set_xlabel("Year")
axes.set_ylabel("CrudeOil")
axes.set_title("trend of crude oil gdp in millions")
plt.legend()
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2,ncols=2,figsize=(20,10))
axes[0,0].bar(df2["Year"],df2["Agriculture"],color="green")
axes[0,0].set_xlabel("Year",color="#44A825", fontsize=12)
axes[0,0].set_ylabel("Agricuture",color="navy",fontsize=12)
axes[0,0].set_title("Head of various sector GDPO in biilion over the span of years",color="#1E3B9F",fontsize=12)

axes[1,0].step(df2["Year"],df2["Education"],color="#15adb0")
axes[1,0].set_xlabel("Year",color ="#a1b80f", fontsize=12)
axes[1,0].set_ylabel("Education",color="navy",fontsize=12)
axes[1,0].set_title("Head of various sector GDPO in biilion over the span of years",color="#1E3B9F",fontsize=12)

axes[0,1].scatter(df2["Year"],df2["CrudeOil"],color="#01112b")
axes[0,1].set_xlabel("Year",color ="#a1b80f", fontsize=12)
axes[0,1].set_ylabel("CrudeOil",color="navy",fontsize=12)
axes[0,1].set_title("Head of various sector GDPO in biilion over the span of years",color="#1E3B9F",fontsize=12)

axes[1,1].hist(df2["ICT"],bins="auto",color="blue",)
axes[1,1].set_xlabel("Year",color ="#91267e", fontsize=12)
axes[1,1].set_ylabel("ICT",color="navy",fontsize=12)
axes[1,1].set_title("Head of various sector GDPO in biilion over the span of years",color="#1E3B9F",fontsize=12)

plt.show()


```
